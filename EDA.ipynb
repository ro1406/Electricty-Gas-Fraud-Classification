# Plan:
Try to summarize the invoice details about each client. Hence, we have a single row detailing the invoices by a single client. We can then merge the tables (Client and Invoice)              

# Summary statistics on different columns in Invoice.csv

client_id:                  Keep it for merging                      
invoice_date:               Drop it, since we cant summarize dates easily                
tarif_type:                 Mode                
counter_number:             Drop it                
counter_statue:             Drop it                
reading_remarque:           Min and Mode                
counter_coefficient:        Mode (1 and 0 only)                
consommation_level_1:       Mean and Std Div                
consommation_level_2:       Mean and Std Div                 
consommation_level_3:       Mean and Std Div                
consommation_level_4:       Mean and Std Div                
old_index:                  Drop it                
new_index:                  Drop it                
counter_type:               Mode                                
diff_index:                 Mean and Std Div                

```python
invoice=pd.read_csv("./train/invoice_train.csv")
```

```python
def summarizeInvoices(df):
    ''' df is the dataframe with the invoice data, in the same format as supplied'''
    ans=pd.DataFrame({'client_id':df.client_id.unique()})
    
    #Find the diff in index:
    ans['diff_index']=df['new_index']-df['old_index']
    df['diff_index']=df['new_index']-df['old_index']
    
    # Make counter coefficent and type 0 and 1 only:
    df['counter_statue']=df.counter_statue.apply(lambda x:int(x) if x!='A' else 0)
    df['counter_type']=df.counter_type.apply(lambda x: 0 if x=='ELEC' else 1)
    
    print("Calculating the modes...")
    # All Mode values:
    colsToMode=['tarif_type','reading_remarque','counter_coefficient','counter_type']
    for col in colsToMode:
        ans[col+'_mode']=[x if isinstance(x,(int,np.int64)) else x[0] for x in 
                          df.groupby(by='client_id')[col].agg(pd.Series.mode).to_list()]
        
    print("Calculating the Means and Std div...")
    # All Mean and std values:
    colsToMean=['consommation_level_1','consommation_level_2','consommation_level_3','consommation_level_4','diff_index']
    for col in colsToMean:
        ans[col+'_mean']=df.groupby(by='client_id')[col].agg(pd.Series.mean).to_list()
        ans[col+'_stddiv']=df.groupby(by='client_id')[col].agg(np.std).to_list()
    
    print("Calculating the Min for reading_remarque...")
    # Min of reading_remarque:
    ans['reading_remarque_min']=[x if isinstance(x,(int,np.int64)) else x[0] for x in
                                df.groupby(by='client_id')['reading_remarque'].agg(pd.Series.min).to_list()]
    
    return ans
```

```python
def combineDataset(client,invoice):
    '''Client is directly read from original file
    Invoice is preprocessed with summarizeInvoices() function'''
    
    #convert to datetime objects
    client['creation_date']=pd.to_datetime(client['creation_date'])
    
    #We expect the region codes to have geographical meaning, in terms of the values assigned (Eg: 100 to 199 may be of the same region)
    client['region']=client['region'].apply(lambda x: int(str(x)[0]))
    
    finaldf=client.merge(newdf,on='client_id',how='left')
    return finaldf
```

```python
res=combineDataset(client,newdf)
res
```

```python
sns.heatmap(res.isna())
```

We notice only std div columns have some na values. We realize that some std div are nan since those clients only have one sample (hence the std div formula divides by 0), so we shall replace them with 0

```python
res=res.fillna(0)
```

Next, we split the day,month,year out of the date and drop the date column. Moreover, client_id is unique for each attribute, and hence must be dropped since it doesnt directly determine fraud

```python
res['CreationDay']=res['creation_date'].apply(lambda x:int(x.day))
res['CreationMonth']=res['creation_date'].apply(lambda x:int(x.month))
res['CreationYear']=res['creation_date'].apply(lambda x:int(x.year))
res.drop('creation_date',axis=1,inplace=True)
res.drop('client_id',axis=1,inplace=True) 
```
